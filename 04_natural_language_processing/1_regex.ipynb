### 1.1 Regex para URLs

````
with open('1.json') as f:
    data = list(json.load(f))
print(data)
````

````
re_web = re.compile(r"http?s?://[^ ,']*")
````

### 1.3 Regex para direcciones de Bogot√°

````
cases = [
    "La direccion es carrera 33 #77B - 38",
    "Estoy en calle 77A #38C - 93",
    "Antes vivia en avenida 68 #69d - 77"
]
````

````
re_address = re.compile(r'[a-zA-Z]+? \d+[a-zA-Z]? #\d+[a-zA-Z]+? - \d+')
````

````
for case in cases:
    print(case)
    print("    Direcciones --> ", re.findall(re_address, case), "\n")
````

### 1.4 Regex para matr√≠culas de carros

````
cases = [
    "La matricula es RKO123",
    "mi carro tiene pico y placa hoy, la placa es: XZY388",
    "Me molestaron por la placa de mi carro: SEX069"
]
````

### 1.5 Regex para fechas (extraer d√≠a, mes, etc)

## 2. Preprocesamiento

````
nlp = spacy.load('es_core_news_sm')
````

````
def text_preprocessing(text, nlp):
    """Preprocess string to clean stopwords and normalize text."""
    text = unidecode(text).lower()
    removed = re.sub("[^A-Za-z0-9 ]","", text) 
    removed = re.sub(r'\s+', ' ', removed)
    doc = nlp(removed)
    
    # tokenization
    tokens = [token for token in doc]
    
    # remove stop words
    text_no_sw = [token for token in tokens if token.is_stop == False]
    
    # filter tokens (char < 3 and char > 15)
    filtered = [token for token in text_no_sw 
                if len(token.text) >=3 and len(token.text)<=15]
    # lemmas 
    lemmas = [token.lemma_ for token in filtered]

    return { 
        'removed': removed, 
        'tokens': tokens, 
        'no_stop': text_no_sw, 
        'filtered': filtered, 
        'lemmas': lemmas
    }
````
